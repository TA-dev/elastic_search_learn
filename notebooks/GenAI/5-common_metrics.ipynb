We generally divide metrics into three categories: **Overlap**, **Semantic**, and **Diversity**.

---

## 1. N-Gram Overlap Metrics (The "Keyword" Matchers)
These are the most common because they are fast and easy to calculate. They check how many words the AI and the Human have in common.

* **ROUGE (Recall-Oriented Understudy for Gisting Evaluation):**
    * **Focus:** It measures **Recall** (how much of the human summary did the AI manage to cover?).
    * **ROUGE-1:** Matches of individual words.
    * **ROUGE-2:** Matches of two-word pairs (bigrams).
    * **ROUGE-L:** The "Longest Common Subsequence." This is the gold standard for summarization because it rewards the model for keeping words in the correct order.
* **BLEU (Bilingual Evaluation Understudy):**
    * **Focus:** Originally for translation. It measures **Precision** (of the words the AI generated, how many were actually in the human version?).
    * **The Catch:** It’s very strict. If the AI uses a perfect synonym (e.g., "happy" instead of "glad"), BLEU gives it a 0 for that word.



---

## 2. Semantic/Embedding Metrics (The "Meaning" Matchers)
These use AI models (like the `SentenceTransformer` you just used) to understand the *vibe* of the text, even if the words are different.

* **Cosine Similarity:**
    * As you saw, it converts sentences into vectors in a high-dimensional space.
    * **Logic:** It measures the angle between the vectors. Smaller angle = more similar meaning.
* **BERTScore:**
    * Instead of looking at the whole sentence as one vector, BERTScore aligns every word in the AI summary with the most similar word in the Human summary. 
    * It’s excellent at detecting if the "gist" is correct even with total paraphrasing.



---

## 3. Linguistic & Diversity Metrics (The "Quality" Checkers)
These don't care about the "Human Reference"—they only look at the AI's output to see if it's "garbage."

* **Perplexity (PPL):**
    * **Focus:** How "confused" is the model by the text?
    * **Logic:** A lower perplexity means the text is fluent and follows the rules of grammar. A high perplexity means the model is spitting out gibberish.
* **Self-BLEU / Diversity Score:**
    * This checks if the model is repeating itself. If you generate 10 summaries and they are all 99% identical, your Diversity Score will be very low (a bad sign).
* **Compression Ratio:**
    * Specific to summarization. It’s simply: $Length(Summary) / Length(Article)$.
    * If this is close to 1.0, your model isn't summarizing; it's "echoing."
