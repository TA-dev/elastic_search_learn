# Generative AI Experiments

This repository contains a collection of Jupyter Notebooks focused on generative AI.
The projects explore **multiple data types** — text, images, and audio — with independently defined goals, datasets, and model choices.

While initially inspired by tutorials *in terms of data modalities*, all implementations and extensions are original and tailored.

---

## Project Structure

### `text_to_image.ipynb`

Implementation of a text-to-image generation pipeline.

* **Goal**: Train/fine-tune a *student model* on a new visual style using high-quality outputs from DiffusionDB
* **Key Features**:

  * Exploratory Data Analysis (EDA) of DiffusionDB metadata (e.g., CFG Scale, Sampling Steps)
  * Just-In-Time (JIT) data loading pipeline for handling large datasets with minimal RAM usage
* **Core Libraries**: `diffusers`, `transformers`, `accelerate`, `kagglehub`

---

### `text_summarizer.ipynb`

An end-to-end NLP pipeline for automated text summarization.

* **Architecture**: Encoder–Decoder Transformer

  * Encoder extracts key information from long text
  * Decoder generates summaries sequentially
* **Training Setup**:

  * Cross-Entropy Loss
  * AdamW optimizer

## Speech-to-Text (STT) Module

This folder contains a full pipeline for building and fine-tuning an Automatic Speech Recognition (ASR) system using **Wav2Vec2**, adapted for domain-specific technical audio.

### Contents

#### `1-ASR_intro_Wav2Vec.ipynb`

* Introduction to ASR concepts
* Explanation of Wav2Vec2 architecture

#### `2-STT_EDA_explanations.ipynb`

* Dataset exploration and structure
* Text normalization (technical transcripts)
* Audio preprocessing (resampling, durations)
* Input distribution analysis

#### `3-EDA_code.ipynb`

* Efficient preprocessing implementations
* Memory-aware dataset handling (JIT-style)

#### `4-training.ipynb`

* Fine-tuning Wav2Vec2 on domain-specific audio
* Validation-based evaluation
* Metrics tracking (Loss, WER)

---

## Model Setup (STT)

* **Base Model**: Wav2Vec2 pretrained on LibriSpeech (960h)
* **Adaptation**: LoRA + classification head
* **Trainable Parameters**: ~1M
* **Objective**: Improve transcription accuracy for long, technical recordings
