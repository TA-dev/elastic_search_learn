# Generative AI & NLP Experiments
This repository contains a collection of Jupyter Notebooks focused on Natural Language Processing (NLP) and Computer Vision using the Hugging Face ecosystem.

## Project Structure
* **text_to_image.ipynb**: Implementation of a text-to-image pipeline.
    * **Goal**: Training/Fine-tuning a "Student" model (SDXL-Turbo) to match high-quality outputs from the DiffusionDB dataset.
    * **Key Features**: Includes Exploratory Data Analysis (EDA) of DiffusionDB metadata (CFG Scale, Sampling Steps) and a Just-In-Time (JIT) data loading pipeline to handle massive datasets with low RAM usage.
    * **Core Libraries**: diffusers, transformers, accelerate, kagglehub.

* **text_summarizer.ipynb**: An end-to-end NLP pipeline for automated text summarization.
    * **Architecture**: Uses an Encoder-Decoder transformer architecture. The Encoder extracts the essence of long articles, while the Decoder generates summaries word-by-word.
    * **Training Specs**: Utilizes Teacher Forcing and Cross-Entropy Loss for optimization with the AdamW optimizer.
