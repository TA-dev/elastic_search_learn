# What is Google ADK?
open-source source python toolkit for building, evaluating and deploying AI agents.
Gives fine-grained control with integration to Google Cloud.
A clean and straightforward way of building agents.

Google ADK provides:
A. LLM-Based Agents: reasoning, tools, transfer.
B. Workflow Agents: SequentialAgent, ParallelAgent, Loop Agent.
C. Custom Logic: CustomAgent.

**Note**: all the agents inherit from and extend the BaseAgent class.

# Getting Started
1. Go to Google CLoud console and create a new project.
2. Select the project in our console (go to the notification part to select).
3. Head over to Google AI studio.
4. Create API key (sign int -> "Get API key" button on top right corner of the screen).
5. Copy the API key and save it in our '.env' file.
6. Download the following requirements:\

    litellm==1.66.3\
