# Lesson 1: Components of an AI Agent

1. **LLM (Large Language Model)**

   * Provides the reasoning behind the AI agent.
   * Identifies the task and the steps needed to complete it.
   * Performs or guides the execution of those steps.

2. **Memory**

   * **Short-term:** Maintains context during the current conversation.
   * **Long-term:** Stores knowledge and experiences over time to improve performance.

3. **Tools**

   * APIs, services, or functions that the agent can access.
   * Provides data or capabilities to help determine and execute actions.

**→ Ultimately, the AI agent uses these components to perform a task efficiently.**

# Lesson 2: Agentic Frameworks

* **Purpose:** Tools that help manage and control tasks effectively.
* **Multi-Agent Support:** Some tasks require multiple agents, and frameworks allow us to coordinate them.
* **Resource Management:** Frameworks help keep track of the state of resources across agents.
* **Agent Collaboration:** They define how agents work together to complete tasks, supporting multiple protocols and workflows.
* **Examples of Frameworks:**

  * **Azure AI:** Designed for single-agent setups.
  * **Semantic Kernel:** Enterprise-focused framework.
  * **AutoGen:** Research-oriented; can integrate with Azure and GitHub agents.

# Lesson 3: What Makes a Good AI Agent?

1. **Space**:
   The environment in which the agent operates, connecting events, people, and knowledge. It should be easily discoverable and allow smooth transitions between foreground and background activities. Achievable through a well-designed UI that clearly explains the agent’s capabilities, usage instructions, and limitations.

2. **Time**:
   How the agent interacts with the user over time. A good agent can improve through repeated interactions if designed properly, for example, by leveraging long-term memory.

3. **Core**:
   Embracing uncertainty and enabling trust and transparency. This can be achieved by providing users with diverse controls and clear access to the agent’s decision-making process.

# Lesson 4: Agent Tool Use Design Pattern

**Agentic Design Patterns** are architectural and behavioral templates that guide the development of AI systems. They define how agents interact with **tools, environments, and sometimes each other**, enabling scalable, modular, and adaptable designs.

Key principles for agent design include:

* **Efficiency** – minimize latency and maximize throughput.
* **Robustness** – resilience to system failures.
* **Scalability** – handle an increasing number of tasks effectively.

## Common Agentic Design Patterns

1. **Reflection Pattern**

   * Agents introspect and evaluate their actions or decisions for iterative improvement.
   * **Design Flow:** Perform action → collect results → compare against predefined goals → refine strategy/logic.

2. **Tool Use Pattern**

   * Agents leverage external tools or APIs to extend their capabilities.
   * Useful for delegating specialized tasks that the agent cannot perform natively.

3. **Planning Pattern**

   * Agents formulate and execute multi-step plans to achieve complex objectives.
   * Plans are dynamically adapted based on goals and constraints.

4. **Multi-Agent Pattern**

   * Multiple agents coordinate to solve complex tasks collaboratively.
   * Useful when tasks require diverse skills or parallel execution.

# Lesson 5: Agentic RAG vs. Basic RAG
## What is RAG?
**RAG (Retrieval-Augmented Generation)** is a technique that enhances Large Language Models (LLMs) by allowing them to access external data sources.
Instead of relying only on their training data, the model:

1. Receives a user query
2. Retrieves relevant information from a database or knowledge source
3. Injects that information into its context
4. Generates a more accurate and up-to-date response

## What is Agentic RAG?
**Agentic RAG** extends basic RAG by introducing decision-making and iterative reasoning.
Instead of performing a single retrieval step, an agent:

1. **Analyzes the query** and breaks it into smaller sub-tasks
2. **Selects tools and data sources** dynamically based on the task
3. **Retrieves and evaluates information**
4. **Checks whether the answer is sufficient**
5. **Repeats the process if needed** (iterative loop)

This creates a more flexible and intelligent system that can handle complex queries.

### Key Advantages of Agentic RAG

* **Multi-step reasoning** – breaks complex problems into manageable parts
* **Tool usage** – can call different APIs, databases, or functions
* **Self-evaluation** – verifies whether the retrieved information is sufficient
* **Iterative improvement** – refines results through multiple passes
* **Memory** – can retain past attempts and avoid repeating ineffective steps

# Lesson 6: How to Build Effective AI Agents
## 1. System Message Framework

* The **system message** provides the highest level of control over an AI agent.
* Defines the agent’s **behavior, goals, constraints, and tone**.
* Ensures the agent performs the **intended actions consistently**.

### Framework Approach

* Instead of writing prompts manually:

  1. Start with a **simple system message**
  2. Use an LLM with a **prompt template**
  3. Generate a **refined and improved system message**

* This enables:

  * **Scalability** – reuse across multiple agents
  * **Consistency** – standardized behavior
  * **Iteration** – continuous improvement over time


## 2. Iterative Prompt Engineering

* Prompt design is an **iterative process**.
* Small improvements to the template lead to **better system messages across all agents**.
* The framework allows:

  * Easy **tweaking and experimentation**
  * Gradual **performance optimization**
  * More **reliable and predictable agent behavior**


## 3. Human-in-the-Loop Architecture

* Keeps humans involved in **critical decision-making steps**.
* Agents are instructed to wait for **specific user signals** before acting.

### Example

* User says **"approve"** → agent proceeds with action
* No approval → agent **pauses, asks, or stops execution**

### Why It Matters

* **Safety** – prevents unintended or harmful actions
* **Control** – user validates important steps
* **Accountability** – decisions remain human-supervised


**→ Effective agents are built through structured prompts, continuous iteration, and controlled execution with human oversight.**


# Lesson 7: Planning Design Pattern

## 1. What is the Planning Design Pattern?

* The **Planning design pattern** enables an AI agent to break down a complex task into smaller, manageable subtasks.
* Instead of solving everything at once, the agent creates a **step-by-step plan**.

### Example

* Task: Plan a 3-day vacation

  * Flights
  * Hotels
  * Transportation
  * Activities

* Each subtask can be handled independently.

## 2. How Does It Help AI Agents Complete Tasks?

* **Decomposition** – breaks complex problems into simpler parts

* **Clarity** – defines a clear sequence of actions

* **Efficiency** – reduces confusion and improves execution

* **Error Handling** – easier to detect and fix issues in smaller steps

* Helps agents move from:

  * “Solve everything at once”
    → to
  * “Solve step-by-step with structure”

## 3. Using Planning with Multiple Agents

* Planning becomes more powerful in **multi-agent systems**.

* Each subtask can be:

  * Assigned to a **different agent**
  * Executed as a **separate process or tool call**

### Example Flow

1. Main agent creates a plan
2. Subtasks are distributed:

   * Agent A → Flights
   * Agent B → Hotels
   * Agent C → Activities
3. Results are combined into a final output

## 4. How Can We Use It in Our Agents?

* Add a **planning step before execution**

* Store the plan as structured output (list of tasks)

* Execute tasks:

  * Sequentially (step-by-step)
  * Or in parallel (multi-agent setup)

* Optionally include:

  * **Re-planning** if a step fails
  * **Evaluation** after each step

**→ Planning enables agents to solve complex problems by structuring them into clear, executable steps.**


# Lesson 8: How to Use a Multi-AI Agent System

## 1. What is a Multi-Agent System?

* A **multi-agent system** consists of multiple AI agents working together to complete a task.
* Each agent has a **specific role or specialization**.

### Core Components

* **Group Chat**

  * Messages are **broadcast to all agents**
  * Enables shared context and communication

* **Group Chat Manager**

  * Typically another agent
  * Selects the **appropriate agent to act next**
  * Coordinates the workflow

## 2. Common Interaction Patterns

### Handoff Pattern

* One agent transfers control to another agent when needed.
* Used when a task requires **information or expertise from another agent**.

### Collaborative Filtering

* Multiple agents contribute **different perspectives or analyses**.

* Each agent acts as a **specialist** in a specific domain.

* Useful for:

  * Analysis tasks
  * Decision-making
  * Generating diverse outputs


## 3. Example of a Multi-Agent Workflow

* Agent A → Generates content

* Agent B → Evaluates and provides feedback

* Workflow:

  * **Generation → Evaluation → Improvement**

## 4. Multi-Agent vs. Single-Agent Systems

### Multi-Agent

* **Pros:**

  * Specialized agents → better performance per task
  * Parallel execution → faster workflows
  * Diverse perspectives → richer outputs

* **Cons:**

  * More complex coordination
  * Higher resource usage
  * Requires management (e.g., chat manager)

### Single-Agent

* **Pros:**

  * Simpler design
  * Easier to manage and debug
  * Lower overhead

* **Cons:**

  * Limited specialization
  * Harder to scale for complex tasks


## 5. Control in Multi-Agent Systems

* Control is achieved through:

  * **Agent roles** – clear responsibilities
  * **Manager agent** – directs task flow
  * **Communication structure** – how agents share information

* Important considerations:

  * Avoid unnecessary loops or conflicts
  * Ensure clear task ownership
  * Maintain efficiency and coherence

## 6. Termination Conditions

* **Termination functions** define when a multi-agent interaction should stop.
* Prevents infinite loops and unnecessary computation.

### Why It Matters

* **Efficiency** – avoids wasting resources
* **Stability** – prevents endless back-and-forth between agents
* **Clarity** – defines when a task is considered complete

### **Example: Writer–Checker System**

* Agent A → Writes content

* Agent B → Checks and evaluates

* Workflow:

  * Writer generates content
  * Checker reviews
  * If **not approved** → send feedback → Writer revises
  * If **approved** → **terminate the conversation**

### **Common Termination Signals**

* Explicit keyword (e.g., **"approved"**, **"done"**)
* Confidence threshold reached
* Maximum number of iterations
* Task completion criteria met
