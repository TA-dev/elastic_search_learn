*Note: ChatGPT is not a standalone model. It was trained internally using an LLM (GPT-3.5 or GPT-4).*

1. **Generative Pre-training**  
   - Trained on a large corpus of text using transformer architecture.  
   - Learns text embeddings and general language patterns.  

2. **Supervised Fine-Tuning (SFT)**  
   - Human AI trainers provide example conversations, acting as both the human and AI assistant.  
   - Used for supervised learning.  
   - Question: What happens if the input is not in the training data?  
     - Answer: Reinforcement learning helps adapt.  
     

3. **Reinforcement Learning**  
   - Fine-tuned against a reward model that scores outputs for alignment with desired behavior.
