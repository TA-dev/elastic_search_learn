# Speech to Text (ASR)
* Interpret and transcribe spoken language into written text.
* ASR systems can be employed for speech synthesis, generating spoken language from text or other inputs.

## 2 Types of ASR Systems
1. **Rule-Based ASR**:  predefined rules and patterns to match acoustic features with linguistic units.
   * Effective in well-defined language and low-noise scenarios.
   * Restricted vocabulary.
2. **Statistical ASR**: utilizes statistical models to map acoustic features to linguistic units.
   * Substantial training data and computational resources for accurate models.
   * Well-suited for applications dealing with varying vocabulary sizes and common challenges like ambient noise.
3. **Neural Network-Based ASR**: utilizes deep learning to achieve high accuracy and robust performance in transcribing speech.
   * Excels in recognizing nuances.
   * Requires substantial training data and computational resources.
   * For applications prioritizing precision and adaptability in handling diverse linguistic patterns

     

         * Transformer-Based ASR
         * CNN-Based ASR
         * Hybrid Neural Network-Based ASR


## Wav2Vec: Self-Supervised Speech Learning
* **Contrastive learning**: the model learns to identify the correct latent speech unit at a given time step among distractors. Achieved by maximizing similarity between:
    * Positive pair: contextualized representation and its matching quantized latent vector at the same time step.
    * Negative pairs: contextualized representation and quantized vectors from other time steps.
### Components:
* Speech Encoder:
    * Multi-layer CNN encoder (speech to latent speech representation: speech characteristics).
    * Transformer: (latent speech to contextualized representation: capturing dependencies). -> Further processed by a context network with transformer architecture.
    * Quantization Module: (quantization of latent speech representation, using codebooks chosen by Gumnel-Softmax). 
* Contrastive Loss Function: betweencontextualized representaion and the qunatization result (the target).
